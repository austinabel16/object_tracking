```python
import cv2
from IPython.display import display, Image, clear_output
import time

# Path to the video file
video_path = 'highway.mp4'  # Replace with your video file path

# Open the video file
cap = cv2.VideoCapture(video_path)

# Check if the video file opened successfully
if not cap.isOpened():
    print("Error: Could not open video file.")
    cap.release()
    exit()

# Video Playback in Jupyter Notebook
try:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            print("End of video or error reading the file.")
            break

        # Encode frame as a JPEG image for display
        _, buffer = cv2.imencode('.jpg', frame)
        img_bytes = buffer.tobytes()
        
        # Display the frame in the notebook
        clear_output(wait=True)
        display(Image(data=img_bytes))
        
        # Adjust playback speed
        time.sleep(1 / cap.get(cv2.CAP_PROP_FPS))

except KeyboardInterrupt:
    print("Video playback interrupted by user.")

# Release resources
cap.release()
cv2.destroyAllWindows()
```
